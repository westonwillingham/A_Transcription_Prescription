```python
text = 'Now is the time for all good men to come to the aid of their country'
```

```python
def data_generator(text):
    while True:
        y = np.random.choice(text.lower().split())
        plt.text(1,1, y, size = 'xx-large')
        plt.xlim([0,2])
        plt.ylim([0,2])
        plt.axis('off')
        plt.savefig(f'{y}.jpg')
        plt.close()
        X = np.array(Image.open(f'{y}.jpg'))
        yield X, y
```
